```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3, 3) # it creates a quantum circuit with 3 quantum bits and 3 classical bits
```

```python
%matplotlib inline
circuit.draw(output = 'mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output = 'mpl')
```

```python
# Criando um emaranhamento entre o qubit 1 e 2 
circuit.h(1)
circuit.cx(1,2)
circuit.draw(output = 'mpl')
```

```python
# Next teletransportation Protocol
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
circuit.measure([0,1],[0,1]) # medindo os qubits 0 e 1 e colocando nos bits 0 e 1
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output = 'mpl')
```

```python
# Assim, a informao que estava no qubit 0 passou para o qubit 2
circuit.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
